**Question 1. A die is rolled. What is the probability of getting:**

**(a) An even number**

**(b) A number greater than 4**

- Total possible outcomes when a die is rolled:
S = {1, 2, 3, 4, 5, 6}
n(S) = 6

Probability Formula:
P(E) = Number of favorable outcomes / Total number of outcomes


(a) Probability of getting an even number

Even numbers on a die = {2, 4, 6}

Number of favorable outcomes = 3

P(Even number) = 3 / 6
P(Even number) = 1 / 2
P(Even number) = 0.5

Answer: 1/2


(b) Probability of getting a number greater than 4

Numbers greater than 4 = {5, 6}

Number of favorable outcomes = 2

P(Number > 4) = 2 / 6
P(Number > 4) = 1 / 3
P(Number > 4) ≈ 0.33

Answer: 1/3

**Question 2: In a class of 50 students:**

**20 like Mathematics (M)**

**15 like Science (S)**

**5 like both subjects**

**What is the probability that a student chosen at random likes Mathematics or Science?**

 - Number of students who like Mathematics (M) = 20  
Number of students who like Science (S) = 15  
Number of students who like both subjects (M ∩ S) = 5  
Total number of students = 50

Formula:
P(M ∪ S) = [ n(M) + n(S) − n(M ∩ S) ] / Total students

Substitute the values:

P(M ∪ S) = (20 + 15 − 5) / 50  
P(M ∪ S) = 30 / 50  
P(M ∪ S) = 3 / 5  
P(M ∪ S) = 0.6

Final Answer:
The probability that a randomly chosen student likes Mathematics or Science = **3/5 (or 0.6)**.

**Question 3: A bag has 3 red and 2 blue balls. If one ball is drawn randomly and is red, what is the probability

- Total balls = 5

Remaining balls in the bag:

Red balls left = 2

P(E) = Number of favorable outcomes / Total number of outcomes

**Question 4: The population of a school is divided into 60% boys and 40% girls. If you want equal representation
of both genders in the sample, which method should you use: Simple Random Sampling or Stratified Sampling?
Why?**
 - The population of the school consists of:
Boys = 60%
Girls = 40%

To ensure that both genders are represented equally in the sample, the appropriate method is **Stratified Sampling**.

Explanation:
In Stratified Sampling, the population is divided into different groups called strata based on specific characteristics (here, gender: boys and girls). After dividing the population into these groups, samples are selected from each group separately.

This method ensures that both boys and girls are properly represented in the sample.

Why not Simple Random Sampling?
In Simple Random Sampling, every student has an equal chance of being selected, but it does not guarantee equal representation of boys and girls. One group might be overrepresented while the other might be underrepresented.

Hence,
The appropriate method is **Stratified Sampling** because it ensures equal representation of both boys and girls in the sample.

- Average height of population (μ) = 160 cm

**Question 8: From a deck of 52 cards, two cards are drawn without replacement. What is the probability that
both are Aces?**

- From a deck of 52 cards, two cards are drawn without replacement.

Total cards in a deck = 52  
Number of Aces in the deck = 4  

Step 1: Probability that the first card drawn is an Ace

P(First Ace) = 4 / 52

Step 2: After one Ace is drawn (without replacement)

Remaining cards = 51  
Remaining Aces = 3  

P(Second Ace | First Ace) = 3 / 51

Step 3: Multiply the probabilities

P(Both Aces) = (4/52) × (3/51)

P(Both Aces) = 12 / 2652

P(Both Aces) = 1 / 221

P(Both Aces) ≈ 0.00452

Hence,
The probability that both cards drawn are Aces = **1/221 (≈ 0.00452)**.

**Question 9: A factory produces bulbs with 2% defective rate. If 5 bulbs are chosen at random, what is the
probability that all are non-defective?**

 - A factory produces bulbs with a 2% defective rate.

Defective bulbs = 2% = 0.02  
Non-defective bulbs = 1 − 0.02 = 0.98

Number of bulbs selected = 5

Probability that one bulb is non-defective = 0.98

Probability that all 5 bulbs are non-defective:

P(All non-defective) = (0.98)^5

P(All non-defective) = 0.9039 (approx)

Hence,
The probability that all 5 bulbs are non-defective ≈ **0.9039** (or **90.39%**).

**Question 10: Differentiate between discrete and continuous random variables with examples.**
 - Differentiate between Discrete and Continuous Random Variables

A **random variable** is a variable whose value depends on the outcome of a random experiment.

1. Discrete Random Variable

Definition:
A discrete random variable can take only **countable and specific values**.

Characteristics:
- Values are usually whole numbers.
- The number of possible outcomes is finite or countably infinite.

Example:
- Number of students present in a class.
- Number of heads obtained when tossing a coin 3 times.
- Number of defective items in a batch.

Example Values:
0, 1, 2, 3, 4, ...


2. Continuous Random Variable

Definition:
A continuous random variable can take **any value within a given range or interval**.

Characteristics:
- Values can include fractions or decimals.
- The number of possible values is infinite within a range.

Example:
- Height of students.
- Weight of a person.
- Temperature of a city.

Example Values:
150.2 cm, 160.5 cm, 170.75 cm, etc.


Key Difference:

Discrete Random Variable:
- Takes countable values.
- Example: Number of cars in a parking lot.

Continuous Random Variable:
- Takes infinite values within a range.
- Example: Height or weight of a person.
